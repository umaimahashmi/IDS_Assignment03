```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Loading the dataset
img1_path = "/content/img1.png"
img2_path = "/content/img2.png"
img3_path = "/content/img3.png"
img4_path = "/content/img4.png"
img5_path = "/content/img5.png"
img6_path = "/content/img6.png"
img7_path = "/content/img7.png"
img8_path = "/content/img8.png"
img9_path = "/content/img9.png"
```

```python
img= cv2.imread(img1_path)                   # read the image using OpenCV
img1= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)     # convert the color from BGR to RGB for displaying using matplotlib

plt.imshow(img1) # display the image using matplotlib
plt.axis('off')
plt.show()
```

```python
img= cv2.imread(img2_path)                   # read the image using OpenCV
img2= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)     # convert the color from BGR to RGB for displaying using matplotlib

plt.imshow(img2) # display the image using matplotlib
plt.axis('off')
plt.show()
```

```python
angle = 285
height, width= img2.shape[:2]     # Get the dimensions of the image

rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1.0)    # Calculate the rotation matrix
img2_rotated= cv2.warpAffine(img2, rotation_matrix, (width, height))             # Apply the rotation to the image

# Display the original and rotated images
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
axs[0].set_title("Original Image")
axs[1].imshow(cv2.cvtColor(img2_rotated, cv2.COLOR_BGR2RGB))
axs[1].set_title("Rotate Image")
axs[0].axis('off')
axs[1].axis('off')
plt.show()
```
